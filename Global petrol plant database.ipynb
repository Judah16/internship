In latitude,longitude,commissioning_year,year_of_capacity_data and generation_gwh_2018 still have missing value          

```python
# filling missing data for continous values with mean
GPP["latitude"].fillna(GPP["latitude"].mean(),inplace=True)
GPP["longitude"].fillna(GPP["longitude"].mean(),inplace=True)

# filling missing data for categorical values with mode
GPP["commissioning_year"].fillna(GPP["commissioning_year"].mode()[0],inplace=True)
GPP["geolocation_source"].fillna(GPP["geolocation_source"].mode()[0],inplace=True)
GPP["generation_gwh_2018"].fillna(GPP["generation_gwh_2018"].mode()[0],inplace=True)
```

```python
GPP.isnull().sum()
```

Since we had to take care of the missing data I have chosen to fill the null values in continous data column with it's mean and the null values for categorical data column with it's mode information
