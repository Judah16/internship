Que 1 ) Scrape the details of most viewed videos on YouTube from Wikipedia.                                                        
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos                                                     
You need to find following details:                                                                                        
A) Rank                                                                                                                     
B) Name                                                                                                                     
C) Artist                                                                                                                   
D) Upload date                                                                                                             
E) Views                                                                                                                   


```python
import pandas as pd
import selenium
from selenium import webdriver
import warnings 
warnings.filterwarnings('ignore')
```

```python
driver=webdriver.Chrome(r"C:\Users\MANISHA JUDHA\Desktop\chromedriver.exe")
```

```python
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/¶"
driver.get(url)
```

```python
# Finding the search button using the xpath
search_button = driver.find_element_by_xpath("//a[@class='external text']")

# Clicking the search button
search_button.click()
```

```python
# Finding the list search button using xpath
list_search = driver.find_element_by_xpath("//ul[@class='mw-search-results']/li[1]/div/a")

# Clicking the list search button
list_search.click()
```

```python
# First, Let's create an empty lists to store the scraping data

Rank = []
Name = []
Artist = []
UploadDate = []
Views = []
```

```python
# Finding the element for rank using xpath
rank = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
for i in rank:
    Rank.append(i.text)

# Finding the element for name using xpath 
name = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')
for i in name:
    Name.append(i.text)

# Finding the element for artist using xpath
artist = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')
for i in artist:
    Artist.append(i.text)

# Finding the element for view using xpath
view = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
for i in view:
    Views.append(i.text)

# Finding the element for date using xpath
date = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
for i in date:
    UploadDate.append(i.text)
```

```python
# Import pandas as pd
ques_1 = pd.DataFrame({})
ques_1["Rank"] = Rank[0:30]
ques_1["Video Name"] = Name[0:30]
ques_1["Artist"] = Artist[0:30]
ques_1["Upload Date"] = UploadDate[0:30]
ques_1["Views (Billions)"] = Views[0:30]
ques_1.index+=1

# Details of most viewed videos on YouTube from Wikipedia
ques_1
```

```python
# First, Let's create an empty lists to store the scraping data

Rank = [] 
State = [] 
GSDP_Year_2019_to_Year_2020 = [] 
GSDP_Year_2018_to_Year_2019 = [] 
Share_Year_2018_to_2019 = [] 
GDP_Dollar_Billion = []
```

```python
# Finding the elements for rank using xpath
rank = driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr/td[1]')
for i in rank:
    Rank.append(i.text)

# Finding the elements for state using xpath
state = driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr/td[2]')
for i in state:
    State.append(i.text)

# Finding the elements for GSDP using xpath
gsdp_19_20 = driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr/td[3]')
for i in gsdp_19_20:
    GSDP_Year_2019_to_Year_2020.append(i.text)

# Finding the elements for GSDP using xpath
gsdp_18_19 = driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr/td[4]')
for i in gsdp_18_19:
    GSDP_Year_2018_to_Year_2019.append(i.text)

# Finding the element for share using xpath 
share = driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr/td[5]')
for i in share:
    Share_Year_2018_to_2019.append(i.text)

# Finding the element for GDP using xpath
gdp_dollar = driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr/td[6]')
for i in gdp_dollar:
    GDP_Dollar_Billion.append(i.text)
```

```python
# Import pandas as pd
ques_4 = pd.DataFrame({})
ques_4["Rank"] = Rank 
ques_4["State"] = State
ques_4["GSDP (Cr INR at Current prices) Year 2019-2020"] = GSDP_Year_2019_to_Year_2020
ques_4["GSDP (Cr INR at Current prices) Year 2018-2019"] = GSDP_Year_2018_to_Year_2019 
ques_4["Share Year 2018-2019"] = Share_Year_2018_to_2019 
ques_4["GDP ($ Billion)"] = GDP_Dollar_Billion
ques_4.index+=1

# Details of State-wise GDP of India from statisticstime.com
ques_4
```

```python
# Importing all the required libraries
import selenium
import pandas as pd
from selenium import webdriver
```

```python
url_6 = "https://www.billboard.com/"
driver.get(url_6)
```

```python
# Finding the element for Hot 100 songs using xpath
hundred_songs = driver.find_element_by_xpath("//nav[@class='header__subnav bg--light']/ul/li[3]/a")

# Clicking the Hot 100 songs
hundred_songs.click()
```

```python
# First, Let's create an empty lists to store the scraping data

SongName = []
ArtistName = [] 
LastWeekRank = [] 
PeakRank = []
WeeksOnBoard = []
```

```python
# Finding the element for song name using xpath
song = driver.find_elements_by_xpath('//span[@class="chart-element__information__song text--truncate color--primary"]')
for i in song:
    SongName.append(i.text)

# Finding the element for artist name using xpath
artist = driver.find_elements_by_xpath('//span[@class="chart-element__information__artist text--truncate color--secondary"]')
for i in artist:
    ArtistName.append(i.text)

# Finding the element for last week rank using xpath
lastweek = driver.find_elements_by_xpath('//div[@class="chart-element__meta text--center color--secondary text--last"]')
for i in lastweek:
    LastWeekRank.append(i.text)

# Finding the element for peak rank using xpath
peak = driver.find_elements_by_xpath('//div[@class="chart-element__meta text--center color--secondary text--peak"]')
for i in peak:
    PeakRank.append(i.text)

# Finding the lement for weeks on board using xpath
weeks = driver.find_elements_by_xpath('//div[@class="chart-element__meta text--center color--secondary text--week"]')
for i in weeks:
    WeeksOnBoard.append(i.text)
```

```python
# Import pandas as pd
ques_6 = pd.DataFrame({})
ques_6["Song Name"] = SongName
ques_6["Artist Name"] = ArtistName
ques_6["Last Week Rank"] = LastWeekRank
ques_6["Peak Rank"] = PeakRank
ques_6["Weeks On Board"] = WeeksOnBoard
ques_6.index+=1

# Details of top 100 songs on billboard.com
ques_6
```
