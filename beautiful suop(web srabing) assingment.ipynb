### 2   Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)  and make data frame.

```python
url=("https://www.imdb.com/list/ls091520106/")
page=requests.get(url)
soup=BeautifulSoup(page.content,'html.parser')

name = soup.find_all("h3", class_="lister-item-header")
movies_name = []           
for i in name:
    for j in i.find_all("a"):
        movies_name.append(j.text)

year = soup.find_all("span",class_="lister-item-year text-muted unbold")
year_of_release = []      
for k in year:
    year_of_release.append(k.text)

      

rating = soup.find_all("div",class_="ipl-rating-star small")

IMDB_rating = []          
for i in rating:
      IMDB_rating.append(float(i.text))

IMDB_top_100=pd.DataFrame({'movies_name':movies_name,'year_of_release':year_of_release,'IMDB_rating':IMDB_rating })        
 
IMDB_top_100
```

### 3) Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of release) and make data frame.

```python
page=requests.get('https://www.imdb.com/india/top-rated-indian-movies/')
soup=BeautifulSoup(page.content)
Indian_movies=[]
for movie in soup.find_all('td',class_="titleColumn"):
    Indian_movies.append(movie.get_text().replace('\n'," "))
    
rating=[]    
for rate in soup.find_all('td', class_="ratingColumn imdbRating"):
     rating.append(rate.get_text().replace('\n',' '))        

data=pd.DataFrame({'Movies':Indian_movies,'Rating':rating})
data.head()
```

### 7) Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content and the code for the video from the link for the youtube video from the post.

```python
url=('https://coreyms.com/')
page=requests.get(url)
soup=BeautifulSoup(page.content )

heading=[]
date=[]
for h in soup.find_all('h2',class_="entry-title"):
     heading.append(h.text)
heading        
        
for d in soup.find_all('time',class_="entry-time"):
    date.append(d.text)

date

content=[]
for c in soup.find_all('div',class_="entry-content"):
    content.append(c.text.replace('\n',''))

    
content 

df=pd.DataFrame({'Heading':heading,'Date':date,'Content':content})
df
```

### 10) Write a python program to scrape first 10 product details which include:                                              (product name) , (price) , (Image URL)         from https://www.bewakoof.com/women-tshirts?ga_q=tshirts . 

```python
page=requests.get('https://www.bewakoof.com/women-tshirts?ga_q=tshirts&sort=popular')   
soup=BeautifulSoup(page.content)

product = []
for p in soup.find_all('div',class_='productCardDetail'):
    product.append(p.text.split())
    
product  


price=[]
for i in soup.find_all('span',class_='discountedPriceText'):
    price.append(i.text.split())
price    
    
image_url=[]
for i in soup.find_all('img',class_='productImgTag'):
    image_url.append(i['src'])
image_url


product =pd.DataFrame({'product name':product,'price':price,'Image URL':image_url})

product

```
